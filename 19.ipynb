```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Sample customer churn dataset: Replace with your actual dataset
data = {
    'Age': [30, 25, 35, 20, 40, 45, 50, 55],
    'Income': [50000, 70000, 60000, 40000, 80000, 90000, 120000, 100000],
    'Transactions': [2, 1, 3, 1, 2, 3, 4, 5],
    'Churn': [0, 1, 0, 1, 0, 1, 0, 1]  # 0: No churn, 1: Churn
}
df = pd.DataFrame(data)

# Define features and target
X = df[['Age', 'Income', 'Transactions']]
y = df['Churn']

# Train the Decision Tree model
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X, y)

# Visualize the decision tree structure
plt.figure(figsize=(10, 6))
plt.title('Decision Tree for Customer Churn Prediction')
plt.show()
```

```python
# Plot scatter plot to visualize relationships
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income'], c=df['Churn'], cmap='coolwarm', edgecolors='k', alpha=0.8, label='Churn')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Relationship between Age, Income, and Churn')
plt.legend()
plt.colorbar(label='Churn (0: No, 1: Yes)')
plt.show()
```
