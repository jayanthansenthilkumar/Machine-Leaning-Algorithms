```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Replace with your actual dataset
data = {'A': [1, 2, None, 4, 5], 'B': ['x', 'y', 'z', 'w', 'x'], 'C': [10.0, 20.0, None, 40.0, 50.0]}
df = pd.DataFrame(data)

# Cleanse dataset: Impute missing values
df['A'].fillna(df['A'].median(), inplace=True)  # Impute missing values in column 'A' with median
df['C'].fillna(df['C'].mean(), inplace=True)   # Impute missing values in column 'C' with mean

# Convert data types if needed
df['C'] = df['C'].astype(float)

print("Cleaned Dataset:")
print(df)


from sklearn.preprocessing import StandardScaler

# Normalize numerical features
scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[['A', 'C']]), columns=['A', 'C'])

# Visualize distributions before and after normalization
plt.figure(figsize=(12, 6))

# Before normalization
plt.subplot(1, 2, 1)
plt.hist(df['A'], bins=5, color='blue', alpha=0.7, rwidth=0.85)
plt.hist(df['C'], bins=5, color='green', alpha=0.7, rwidth=0.85)
plt.title('Before Normalization')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend(['A', 'C'])

# After normalization
plt.subplot(1, 2, 2)
plt.hist(df_normalized['A'], bins=5, color='blue', alpha=0.7, rwidth=0.85)
plt.hist(df_normalized['C'], bins=5, color='green', alpha=0.7, rwidth=0.85)
plt.title('After Normalization')
plt.xlabel('Normalized Values')
plt.ylabel('Frequency')
plt.legend(['A', 'C'])

plt.tight_layout()
plt.show()
```
